С помощью id лигандов парсим sdf файлы лигандов

```python
def converter(lig_name):
    url = "https://files.rcsb.org/ligands/download/"+lig_name+"_ideal.sdf"
    response = requests.get(url)
    with open(f'{lig_name}_ideal.sdf', 'wb') as file:
        file.write(response.content)
    file_name = f'{lig_name}_ideal.sdf'  #определяем имя файла
    sdf_file = Chem.SDMolSupplier(file_name)  #переводим в sdf формат
    #out_file = open(f'{file_name}smiles.txt', "w")
    for mol in sdf_file:    # для каждой молекулы в sdf файле...
        if mol is not None:             # avoiding compounds that cannot be loaded.
            smi = Chem.MolToSmiles(mol)  #to smiles rdkit
            #df_lig_copy[f"{mol}"] = smi   #в копию таблицы
            return smi
    #        out_file.write(f"{smi}\n")
    #out_file.close()
#if __name__ == "__main__":
#converter(FE_ideal.sdf)
#print("Generated SMILES are present in smiles.txt file")
```

Убираем строчки, где отсутствует лиганд или в качестве лиганда представлена вода (HOH)

```python
df_length = df_lig.shape[0]
for n_str in range(df_length):
    ligands_list = df_lig.loc[n_str, 'Ligands']
    while 'HOH' in ligands_list:
        ligands_list.remove('HOH')
    while 'NA' in ligands_list:
        ligands_list.remove('NA')
    while None in ligands_list:
        ligands_list.remove(None)        
    for lig_num in range(len(ligands_list)):
        try:
            ligands_list[lig_num] = converter(ligands_list[lig_num])
        except Exception as e:
            continue 
    if df_lig.loc[n_str, 'Ligands'] == []:
        df_lig.drop(n_str, inplace=True)
```
